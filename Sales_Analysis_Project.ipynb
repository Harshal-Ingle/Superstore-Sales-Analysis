```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Data Cleaning

```python
# 1. Check for missing values
df.isnull().sum()
```

```python
# 2. Check for duplicate rows
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
# 3. Convert date columns from 'object' to 'datetime'
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d/%m/%Y')
```

```python
# 4. Creating useful columns
df['Profit Margin'] = (df['Profit'] / df['Sales']) * 100
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
```

```python
df.head()
```

### Exploration Data Analysis - EDA

##### Q1: What are the sales and profit trends over time?

```python
# Grouping sales by month
monthly_sales = df.groupby(pd.Grouper(key='Order Date', freq='M')).agg({'Sales': 'sum'}).reset_index()

# Plotting monthly sales
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales['Order Date'], monthly_sales['Sales'])
plt.title('Total Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

##### Q2: Which product categories are most profitable?

```python
# We have to Group it by Category and Sub-Category to see the most profitable category
category_profit = df.groupby(['Category', 'Sub-Category']).agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()

# Visualize top 10 profitable sub-categories
top_10_profit = category_profit.sort_values(by='Profit', ascending=False).head(5)

plt.figure(figsize=(10, 8))
sns.barplot(data=top_10_profit, y='Sub-Category', x='Profit', palette='viridis')
plt.title('Top 10 Most Profitable Sub-Categories')
plt.show()
```

##### # Q3: Who are our most valuable customers?

```python
# Group by customer name
customer_value = df.groupby('Customer Name').agg({'Sales': 'sum', 'Profit': 'sum'}).sort_values(by='Sales', ascending=False).head(10)
print("\nTop 10 Customers by Sales:")
print(customer_value)
```

##### Q4: Which regions or states are top performers?

```python
# Group by State
# This will show states by text
state_performance = df.groupby('State').agg({'Sales': 'sum', 'Profit': 'sum'}).sort_values(by='Sales', ascending=False).head(10)
print("Top 10 States by Sales:")
print(state_performance)

# Now by Using Visualization
state_performance.plot(kind='bar', y='Sales', figsize=(12, 6))
plt.title('Top 10 States by Sales')
plt.ylabel('Total Sales')
plt.show()
```

## Summarizing Insights & Recommendations

##### Insight 1: Seasonal Sales Peaks: Sales consistently peak in November and December.

##### Insight 2: Problem with "Tables": The "Tables" sub-category generates significant sales but is a major source of financial loss.

##### Insight 3: High Discounts Hurt Profitability: Discounts above 20-30% consistently lead to a net loss on orders.

##### Insight 4: Geographic Concentration: Sales and profits are heavily concentrated in a few states like California and New York.
